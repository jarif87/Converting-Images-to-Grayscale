```python
import cv2
from PIL import Image
from matplotlib import pyplot as plt
```

```python
image=Image.open("/kaggle/input/custom-image-dataset/cat.jpg")
plt.figure(figsize=(10,8))
plt.imshow(image)
plt.axis('off')  
plt.show()
```

```python
# Load the image using OpenCV
image = cv2.imread("/kaggle/input/custom-image-dataset/cat.jpg")

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using matplotlib
plt.imshow(gray_image, cmap='gray')
plt.axis('off')  # Hide the axes for better visualization
plt.show()
```

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
     # The line below is changed from w, h to h, w
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('/kaggle/input/custom-image-dataset/cat.jpg')

imshow("Castara, Tobago", image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Converted to Grayscale", gray_image)
```
